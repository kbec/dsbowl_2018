```python
import cv2
from glob import glob
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
test_imlist = []
for folder in glob('data/stage1_test/*'):
    image = {}
    image['ImageId'] = folder.split('/')[-1]
    image['Image'] = cv2.imread('{}/images/{}.png'.format(folder, image['ImageId'], cv2.IMREAD_COLOR))
    test_imlist.append(image)
    
test = pd.DataFrame(test_imlist).set_index('ImageId')
test['Image'] = test.Image.apply(np.flip, args=(2,))    # flip order of channels. cv2 reads in BGR, but plt shows RGB
test['ImageShape'] = test.Image.map(lambda a: a.shape)
test.head()
```

```python
train_imlist = []
for folder in glob('data/stage1_train/*'):
    image = {}
    image['ImageId'] = folder.split('/')[-1]
    image['Image'] = cv2.imread('{}/images/{}.png'.format(folder, image['ImageId']), cv2.IMREAD_COLOR)
    train_imlist.append(image)
    
train = pd.DataFrame(train_imlist).set_index('ImageId')
train['Image'] = train.Image.apply(np.flip, args=(2,))
train['ImageShape'] = train.Image.map(lambda a: a.shape)
train.head()
```

```python
train_labels = pd.read_csv('data/stage1_train_labels.csv')
train_labels['EncodedPixels'] = train_labels.EncodedPixels.map(lambda s: [int(x) for x in s.split(' ')])
train_labels.head()
```

```python
def decode_pixels(encoded_list, imshape):
    pix_list = []
    for x in range(len(encoded_list)):
        if x % 2 == 0:
            pix_list.extend([0] * (encoded_list[x] - 1 - len(pix_list)))
        else:
            pix_list.extend([1] * encoded_list[x])
    pix_list.extend([0] * (imshape[0] * imshape[1] - len(pix_list)))
    
    return np.array(pix_list)


def encode_pixels(pix_list):
    pix_list.append(0)
    h = 0
    encoded_list = []
    for i in range(len(pix_list)):
        if pix_list[i] > 0 and h == 0:
            encoded_list.append(i + 1)
        elif pix_list[i] == 0 and h > 0:
            encoded_list.append(i + 1 - encoded_list[-1])
        h = pix_list[i]
        
    return encoded_list
```

```python
def overlay_masks(image_id):
    imlabels = train_labels[train_labels.ImageId == image_id]['EncodedPixels']
    masks = imlabels.apply(decode_pixels, args=(train.ImageShape[image_id],))
    mask_overlay = np.sum(np.stack(masks), axis=0)
    
    return mask_overlay
```

```python
i = 0
```

```python
imid = train.index[i]
print(i, imid)
implot = train.Image[imid]
imshape = train.ImageShape[imid]
imhsv = cv2.cvtColor(implot, cv2.COLOR_RGB2HSV)
imhls = cv2.cvtColor(implot, cv2.COLOR_RGB2HLS)
imlab = cv2.cvtColor(implot, cv2.COLOR_RGB2LAB)


fig, ax = plt.subplots(3, 4, figsize=(20,15))

ax[0,0].imshow(implot)
ax[0,0].set_title('image')
ax[1,0].imshow(overlay_masks(imid).reshape(imshape[1],imshape[0]).T, cmap='gray')
ax[1,0].set_title('mask')
ax[2,0].axis('off')

ax[0,1].imshow(implot[:,:,0], cmap='Reds_r')
ax[0,1].set_title('red')
ax[0,2].imshow(implot[:,:,1], cmap='Greens_r')
ax[0,2].set_title('green')
ax[0,3].imshow(implot[:,:,2], cmap='Blues_r')
ax[0,3].set_title('blue')

ax[1,1].imshow(imhsv[:,:,0], cmap='hsv')
ax[1,1].set_title('hue')
ax[1,2].imshow(imhsv[:,:,1], cmap='nipy_spectral')
ax[1,2].set_title('saturation')
ax[1,3].imshow(imhsv[:,:,2], cmap='gray')
ax[1,3].set_title('value')

ax[2,1].imshow(imlab[:,:,0], cmap='gray')
ax[2,1].set_title('lightness')
ax[2,2].imshow(imlab[:,:,1], cmap='RdYlGn_r')
ax[2,2].set_title('green-red')
ax[2,3].imshow(imlab[:,:,2], cmap='YlGnBu_r')
ax[2,3].set_title('blue-yellow')

i += 1
```
